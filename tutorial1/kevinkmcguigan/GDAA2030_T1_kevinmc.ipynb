```
import numpy as np
from matplotlib import pyplot as plt

# image = plt.imread(r"tutorial1Image.PNG")
# image = plt.imread('/content/GDAA2030/tutorial1/kevinkmcguigan/tutorial1Image.PNG')

import requests
import matplotlib.image as mpimg
url = 'https://github.com/NSCC-COGS/GDAA2030/blob/master/tutorial1/kevinkmcguigan/tutorial1Image.PNG?raw=true'
img_raw = requests.get(url, stream=True).raw
image = mpimg.imread(img_raw)


arr = np.array(image)

Bands = [0,1,2] 

plt.figure(num='kevin k mcguigan')

for index in Bands: 

	plt.subplot(2,3, index +1)
	plt.imshow(arr[:,:,index], cmap="gray")
	plt.subplot(2,3, index+1+3)
	plt.hist(arr[:,:,index].flatten(),bins=20) 

plt.show()
```
